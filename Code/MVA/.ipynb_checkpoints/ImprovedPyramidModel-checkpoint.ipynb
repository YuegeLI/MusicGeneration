```python
from __future__ import print_function
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM, GRU
from keras.optimizers import *

def get_model(batch_size, tsteps, name='Gen', hidden_dim=64, key_num=1):
    print('Creating Model for ', name)
    model = Sequential()
#     model.add(LSTM(hidden_dim,
#                    batch_input_shape=(batch_size, tsteps, key_num),
#                    return_sequences=True,
#                    stateful=True))
    model.add(GRU(hidden_dim,
                   batch_input_shape=(batch_size, tsteps, key_num),
                   return_sequences=False,
                   stateful=True))
    model.add(Dense(key_num))
    optimizer=RMSprop(lr=0.0001)
    model.compile(loss='mse', optimizer=optimizer)
    return model
```

```python
models = []
batch_size = 2
epochs = 400
tsteps = 1 
for i in range(len(Xs)):
    if i == len(Xs) - 1:
        name = 'Gen'
    else:
        name = 'Refine' + str(i+1)
    models.append(get_model(batch_size, tsteps, name=name, key_num=max_key_number))
```

```python
def accuracy (v, v_sample):
    # ACC
    accs = []
    t = len(v)
    n = len(v[0])
    for time in range(t):
        tp = 0 # true positive
        fp = 0 # false positive
        fn = 0 # false negative
        for note in range(n):
            if v[time][note] == 1 and v_sample[time][note] == 1:
                tp += 1.
            if v[time][note] == 0 and v_sample[time][note] == 1:
                fp += 1.
            if v[time][note] == 1 and v_sample[time][note] == 0:
                fn += 1.
        if tp + fp + fn != 0:
            a = tp / (tp + fp + fn)
        else:
            a = 0
        accs.append(a)

    acc = np.mean(accs)
    return acc
    
def cal_accuray(name,model,Current_X):
    accs = []
    for j in range(len(Current_X)):
        y = np.array(model.predict(Current_X[j], batch_size=2))
        model.reset_states()
        target = dt[j]
        if model_type == 'laplacian':
            rst = y.reshape(y.shape[0],1,y.shape[1]) + Current_X[j]
        else:
            rst = y.reshape(y.shape[0],1,y.shape[1])
        sample = frequency2key(rst)
        sample = convert_key_to_midi_list(sample)
        target = frequency2key(target)
        target = convert_key_to_midi_list(target)
        accs.append(accuracy(target,sample))
        break
    acc = np.mean(accs)
    print('ACC for model ', name,': ', acc)    

    
def evulation(name,model,Current_X,Current_y):
    loss = []
    for j in range(len(Current_X)):
        loss.append(model.evaluate(Current_X[j], Current_y[j], batch_size=2, verbose=0))
        model.reset_states()
    print('Avg mse for ' , name , ' in training set is ' , str(sum(loss)*1./len(loss)))
    if name == 'Refine1':
        cal_accuray(name,model,Current_X)
```

```python
from time import time
def training(name, model,epochs, Current_X,Current_y,batch_size):
    print('Training for model: ', name)
    
    for i in range(epochs):
        start = time()
        print('Epoch', i+1, '/', epochs,', for model ', name)
        for j in range(len(Current_X)):
            model.fit(Current_X[j],
                      Current_y[j],
                      batch_size=batch_size,
                      verbose=0,
                      nb_epoch=1,
                      shuffle=False)
            model.reset_states()
        if i % 10 == 0:
            print('loss at epoch', i+1,'for model',name, 'time: ', str(time()-start))
            evulation(name,model,Current_X,Current_y)
    print('Done for', name)

    
# evulation('Refine1',models[0],Xs[0][0],ys[0][0])
# models[0].evaluate(Xs[0][0], ys[0][0], batch_size=1, verbose=0)
# ts = []
# epochs=1
for i in range(len(Xs)):
    Current_X = Xs[i]
    Current_y = ys[i]
    model = models[i]
    if i == len(Xs) - 1:
        name = 'Gen'
    else:
        name = 'Refine' + str(i+1)
    # need to be optimised
    training(name,model,epochs,Current_X,Current_y,batch_size)
#     print('Starting for tread of ', name)
#     t = threading.Thread(target=training, args=(name, model,epochs,Current_X,Current_y,batch_size), name=name)
#     t.start()
#     ts.append(t)
# for t in ts:
#     t.join()
```

```python
def model_predict(model,X,gen_length, predict_type='generator'):
    if predict_type == 'generator':
        y = np.array(model.predict(X, batch_size=2))
        model.reset_states()
        rst = y.reshape(y.shape[0],1,y.shape[1])
        print('generator result shape', rst.shape)
        return rst
    elif predict_type == 'refine':
        y = model.predict(X, batch_size=2)
        rst = y.reshape(y.shape[0],1,y.shape[1]) + X
        print('refine result shape', rst.shape)
        return rst

def generator(models, z=np.array([[[280]*max_key_number]]), gen_length=128):
    gen_model = models[-1]
    if model_type == 'laplacian':
#         gen_model = models[-1]
        generatered = model_predict(gen_model,z,128,predict_type='generator')
        for i in range(len(models)-2, -1, -1):
            refine_model = models[i]
            generatered = model_predict(refine_model,generatered,128,predict_type='refine')
        return generatered
    else:
        generatered = model_predict(gen_model,z,128,predict_type='generator')
        for i in range(len(models)-2, -1, -1):
            refine_model = models[i]
            generatered = model_predict(refine_model,generatered,128,predict_type='generator')
        return generatered
        
```

```python
gen_length = 256
z = Xs[-1][1][0:gen_length]
sample = generator(models,z=z, gen_length=gen_length)
```

```python
import nltk

def get_hypothesis(sample):
    hypothesis = sample.copy()
    hypothesis = frequency2key(hypothesis)
    hypothesis = convert_key_to_midi_list(hypothesis)
    hypothesis = [','.join([str(int(rr)) for rr in r]) for r in hypothesis]
    return hypothesis
    
def get_references(dt):
    references = []
    for target in dt.copy():
        target = frequency2key(target)
        target = convert_key_to_midi_list(target)
        target = [','.join([str(int(rr)) for rr in r]) for r in target]
        references.append(target)
    return references

hypothesis = get_hypothesis(sample)
references = get_references(dt)
#the maximum is bigram, so assign the weight into 2 half.        
```
